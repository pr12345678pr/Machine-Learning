Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing dataset

```python
df=pd.read_csv("Mall_Customers.csv")
x=df.iloc[:,[3,4]].values
```

Using dendogram to find optical number of clusters

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z= linkage(x,method='ward',metric='euclidean')
dendrogram(Z,color_threshold=19)
plt.title("Dendogram")
plt.xlabel("Customers")
plt.ylabel("Eulician Distances")
plt.show()
```

Training the hierarichal clustering model

```python
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_ag=ag.fit_predict(x)
```

```python
print(y_ag)
```

Visualising the clusters

```python
plt.scatter(x[y_ag==0,0],x[y_ag==0,1],s=100,color='red',label='Cluster 1')
plt.scatter(x[y_ag==1,0],x[y_ag==1,1],s=100,color='blue',label='Cluster 2')
plt.scatter(x[y_ag==2,0],x[y_ag==2,1],s=100,color='green',label='Cluster 3')
plt.scatter(x[y_ag==3,0],x[y_ag==3,1],s=100,color='cyan',label='Cluster 4')
plt.scatter(x[y_ag==4,0],x[y_ag==4,1],s=100,c='magenta',label='Cluster 5')
plt.title("Clusters of the customers")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending SCore(1-100)")
plt.legend()
plt.show()
```
